#  6. Cross Validate Spectral Clustering parameters

```python
from sklearn.model_selection import GridSearchCV
from sklearn.cluster import SpectralClustering
from sklearn.metrics import make_scorer, silhouette_score, adjusted_rand_score


param_grid = {
    'n_clusters': [20, 30, 40],  # Adjust based on the number of clusters you want to explore
    'assign_labels': ['kmeans', 'discretize'],
    'affinity': ['nearest_neighbors', 'rbf', 'laplacian']
}

scoring = {
    'adjusted_rand_score': make_scorer(adjusted_rand_score),
    'silhouette_score': make_scorer(silhouette_score, metric='euclidean')
}

# Create the SpectralClustering instance
clustering = SpectralClustering()

# Use GridSearchCV for cross-validation
grid_search = GridSearchCV(clustering, param_grid, cv=5, scoring= scoring, refit='adjusted_rand_score', n_jobs=-1)
```

```python
grid_search.fit(data_fifa23_preproc)

# Print the best parameters
print("Best Parameters:", grid_search.best_params_)

# Get the best model
best_clustering_model = grid_search.best_estimator_

# Fit the best model to the data
best_clustering_model.fit(data_fifa23_preproc)

# Get the labels from the best model
best_labels = best_clustering_model.labels_
print(best_labels)
```
