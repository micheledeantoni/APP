```python
import requests

def country_extraction(country):
    base_url = 'https://restcountries.com/v3.1/name/'
    response = requests.get(base_url+country)
    if response.status_code == 200:
        content = response.json()[0]
        return content['region']
    else:
        return 'Other'
```
